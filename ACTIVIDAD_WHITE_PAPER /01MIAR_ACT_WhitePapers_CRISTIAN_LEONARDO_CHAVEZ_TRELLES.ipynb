```python
import pandas as pd
import matplotlib.pyplot as plt
from prophet import Prophet

df = df_final

# 1. Leer y preparar los datos
df['fecha'] = pd.to_datetime(df['TIEMPO_APERTURA'], errors='coerce')
df = df.dropna(subset=['TIEMPO_APERTURA'])

# Reemplaza con el nombre de tu columna real
valor_col = 'PRECIO_CIERRE'

# 2. Renombrar columnas como exige Prophet
df_prophet = df[['fecha', valor_col]].rename(columns={'fecha': 'ds', valor_col: 'y'})

# 3. Entrenar el modelo
model = Prophet(interval_width=0.95, daily_seasonality=False)
model.fit(df_prophet)

# 4. Crear dataframe futuro para predecir
future = model.make_future_dataframe(periods=0)  # solo predice para los datos disponibles
forecast = model.predict(future)

# 5. Unir predicción con datos reales
df_merged = df_prophet.copy()
df_merged = df_merged.merge(forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']], on='ds')

# 6. Detectar anomalías
df_merged['anomaly'] = (df_merged['y'] < df_merged['yhat_lower']) | (df_merged['y'] > df_merged['yhat_upper'])

# 7. Visualizar resultados
plt.figure(figsize=(14,6))
plt.plot(df_merged['ds'], df_merged['y'], label='Valor real')
plt.plot(df_merged['ds'], df_merged['yhat'], label='Pronóstico', color='green')
plt.fill_between(df_merged['ds'], df_merged['yhat_lower'], df_merged['yhat_upper'],
                 color='lightgreen', alpha=0.4, label='Intervalo de confianza 95%')
plt.scatter(df_merged[df_merged['anomaly']]['ds'], df_merged[df_merged['anomaly']]['y'],
            color='red', label='Anomalías')
plt.title("Detección de Anomalías con Prophet")
plt.legend()
plt.xlabel("Fecha")
plt.ylabel("Valor")
plt.show()
```
